```python
#Importar Pandas
import pandas as pd


#Importar sklearn
import sklearn

#Importar train_test_split para dividir nuestro dataframe
from sklearn.model_selection import train_test_split as tts

#Importar la maquina de soporte de vectores
from sklearn.svm import SVC

#Importar el accuracy
from sklearn.metrics import accuracy_score

#Importar la matriz de confusi√≥n
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('examenm1.csv')
df.head()
```

```python
categoricos = df.iloc[:,[0,3,6,8,11,15]]
categoricos
```

```python
X = df.iloc[:,[1,2,4,7,12,14]]
X
```

```python
y = df.iloc[:,6]
y
```

```python
df.isnull().sum()
```

```python
df = df.fillna({"int_memory": df['int_memory'].mean()})
```

```python
df = df.fillna({"n_cores": df['n_cores'].mean()})
```

```python
df = df.fillna({"talk_time": df['talk_time'].mean()})
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,[1,2,4,7,12,14]]
X
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X,y,test_size=0.2,random_state = 123)
X_train.head()
```

```python
from sklearn.svm import SVC
```

```python
modelo = SVC(C = 100, kernel = 'linear', random_state = 123)
modelo.fit(X_train,y_train)
```

```python
predicted = modelo.predict(X_test)
predicted
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predicted)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predicted)
```
